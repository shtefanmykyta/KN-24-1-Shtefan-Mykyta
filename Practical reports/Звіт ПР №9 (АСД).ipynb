### Відповіді на контрольні питання

Коротко дай відповіді на питання.

1. Що таке кодування Гафмена та як воно працює?
\
Кодування Гафмена — алгоритм стиснення без втрат, де частіші символи отримують коротші коди.

2. Як визначається оптимальний двійковий код Гафмена для стиснення даних?
\
Визначення коду: підрахунок частот символів, побудова бінарного дерева починаючи з найменших частот, присвоєння 0 і 1 гілкам.

3. Які переваги має кодування Гафмена над іншими методами стиснення даних?
\
Переваги: оптимальність для відомих частот, стиснення без втрат, простота реалізації, висока ефективність.

4. Як відбувається декодування даних, закодованих за допомогою кодування Гафмена?
\
Декодування: почергове проходження бітами закодованого повідомлення від кореня дерева до листа (символу).

5. Які є можливі недоліки кодування Гафмена?
\
Недоліки: необхідність зберігати дерево/таблицю кодів, неідеальність при нерівномірних розподілах, статичність.

6. Для чого використовується побудова дерева в кодуванні Гафмена
\
Призначення дерева: створення префіксних кодів, забезпечення однозначності декодування, визначення оптимальних кодів.
